# Steady State Error Analysis

&nbsp;&nbsp;&nbsp; The steady-state error of the system is the value of the error $e_f(t)=r(t)-\lim_{t\to\infty} c(t)$, where $r(t)$ is the reference input and the $c(t)$ is system output. In the steady state error analysis, we will obtain the equation of the steady state error to find out how can we improve the control system.

&nbsp;&nbsp;&nbsp; While we are discussing the servo characteristics of a control system in the Colab notebook of the 6th week, we mentioned the closed loop's error transfer function of the closed-loop (feedback) system is

$$\frac{E(s)}{R(s)}= \frac{1}{1+G_c(s)G(s)H(s)}$$

&nbsp;&nbsp;&nbsp; where $G_c(s)$ is the controller transfer function, $G(s)$ is the plant transfer function and the $H(s)$ is the sensor transfer function. For the simplicity, we will call the open loop transfer function $G_o(s)$ where $G_o(s)=G_c(s)G(s)H(s)$ and the error transfer fucntion can written as

$$\frac{E(s)}{R(s)}= \frac{1}{1+G_o(s)}.$$

&nbsp;&nbsp;&nbsp; Since the $G_o(s)$ is the multiplication of the other transfer functions of the system; it can be described as

$$G_o(s)=\frac{K_{OL}(1+c_1s+...+c_ps^p)}{s^N(1+d_1s+...+d_qs^q)}$$

&nbsp;&nbsp;&nbsp; where $K_{OL}$ is the open-loop gain and the $N$ is called as type number of the system *(which is the number of free integrators, $s$, in the open loop transfer function, $G_o(s)$.)*

&nbsp;&nbsp;&nbsp; From the 7th week Colab, you also know the step input, ramp input, and acceleration inputs, which can be provided as $r(t)$ to the control system. We can formulate them as

$$ r(t)=\frac{r_Mt^M}{M!}h(t) \rightarrow R(s)=\frac{r_M}{s^{M+1}}$$

$$M=0 \rightarrow  r(t)=r_0h(t),\,\,\,step\,\,input$$

$$M=1 \rightarrow  r(t)=r_1th(t),\,\,\,ramp\,\,input$$

$$M=2 \rightarrow  r(t)=r_2\frac{t^2}{2}h(t),\,\,\,acceleration\,\,input$$

&nbsp;&nbsp;&nbsp; Up to now, we remembered the error transfer function and input transfer functions. We will continue with the final value theorem which can be used for finding the steady state error value of the control system.

### Final Value Theorem


&nbsp;&nbsp;&nbsp; The Laplace transform of time function can be described as

$$\mathcal{L}\{f(t)\}= F(s)=\int_{0}^{\infty} f(t)e^{-st} dt$$

&nbsp;&nbsp;&nbsp; and the Laplace transform of $f(t)=\frac{d}{dt}g(t)$ can be found as

$$\mathcal{L}\{\frac{d}{dt}g(t)\}=\int_{0}^{\infty} \frac{d}{dt}g(t)e^{-st} dt = \int_{0}^{\infty} dg(t)e^{-st}=g(t)e^{-st}\Big|_0^\infty+s\int_{0}^{\infty} g(t)e^{-st}dt=sG(s)-g(0).$$

&nbsp;&nbsp;&nbsp; If we take the limit of the Laplace transform 

$$\lim_{s\to 0}\int_{0}^{\infty} \frac{d}{dt}g(t)e^{-st} dt = g(\infty) -g(0) = \lim_{s\to 0} (sG(s)-g(0)), $$

&nbsp;&nbsp;&nbsp; and it is found that

$$\lim_{t\to \infty} g(t) = \lim_{s\to 0} sG(s).$$

&nbsp;&nbsp;&nbsp; By using that equality, we can find the system's final value by taking the limit of the error transfer function $E(s)$ such that 

$$e_f(t)=\lim_{t\to \infty} e(t)= \lim_{s\to 0} sE(s).$$


&nbsp;&nbsp;&nbsp; As we mentioned it earlier the error transfer function is defined as

$$\frac{E(s)}{R(s)}= \frac{1}{1+G_o(s)} \rightarrow E(s) = \frac{1}{1+G_o(s)} R(s)$$

&nbsp;&nbsp;&nbsp; and the input transfer function $R(s)$ can be written as

$$ R(s)=\frac{r_M}{s^{M+1}}.$$

&nbsp;&nbsp;&nbsp; If we substitute the $R(s)$ into the equation of the $E(s)$ 

$$E(s) = \frac{1}{1+G_o(s)} \frac{r_M}{s^{M+1}} = \frac{1}{1+ \frac{K_{OL}(1+c_1s+...+c_ps^p)}{s^N(1+d_1s+...+d_qs^q)}} \frac{r_M}{s^{M+1}}. $$

&nbsp;&nbsp;&nbsp; Then, the steady-state error function can be written as

$$e_f(t)=\lim_{t\to \infty} e(t)= \lim_{s\to 0} sE(s) =\lim_{s\to 0} \frac{1}{1+ \frac{K_{OL}(1+c_1s+...+c_ps^p)}{s^N(1+d_1s+...+d_qs^q)}} \frac{r_M}{s^{M}} = \lim_{s\to 0} \frac {r_M}{s^M+K_{OL}s^{M-N}}.$$

&nbsp;&nbsp;&nbsp; From the obtained expression, we can say that the steady-state error depends on the parameters of the input $R(s)$, type number $N$, and the open-loop gain $K_{OL}$. Now, we are ready to perform the steady-state analysis.

### Step Input


&nbsp;&nbsp;&nbsp; For the step input, $M=0$ in the $R(s)$ and the steady-state error function becomes

$$e_f(t)=\lim_{s\to 0} \frac {r_0}{s^0+K_{OL}s^{0-N}} =\lim_{s\to 0} \frac {r_0}{1+K_{OL}s^{-N}}.$$

&nbsp;&nbsp;&nbsp; Then the steady-state error can be found as

$$
  \begin{align*}
 e_f=
  \begin{cases}
   \frac{r_0}{1+K_{OL}}        & \text{if } N=0. \\
   0        & \text{if } N\geq1.
  \end{cases}
  \
\end{align*}
$$


### Ramp Input


&nbsp;&nbsp;&nbsp; For the ramp input, $M=1$ in the $R(s)$ and the steady-state error function becomes

$$e_f(t)=\lim_{s\to 0} \frac {r_1}{s^1+K_{OL}s^{1-N}} = \lim_{s\to 0} \frac {r_1}{s+K_{OL}s^{1-N}}.$$

&nbsp;&nbsp;&nbsp; Then the steady-state error can be found as

$$
  \begin{align*}
 e_f=
  \begin{cases}
   \infty        & \text{if } N=0. \\
    \frac{r_1}{K_{OL}}        & \text{if } N=1. \\
  0        & \text{if } N\geq2.
  \end{cases}
  \
\end{align*}
$$

### Acceleration Input


&nbsp;&nbsp;&nbsp; For the ramp input, $M=2$ in the $R(s)$ and the steady-state error function becomes

$$e_f(t)=\lim_{s\to 0} \frac {r_2}{s^2+K_{OL}s^{2-N}}.$$

&nbsp;&nbsp;&nbsp; Then the steady-state error can be found as

$$
  \begin{align*}
 e_f=
  \begin{cases}
   \infty        & \text{if } N=0. \\
    \infty        & \text{if } N=1. \\
    \frac{r_2}{K_{OL}}        & \text{if } N=2. \\
  0        & \text{if } N\geq3.
  \end{cases}
  \
\end{align*}
$$

&nbsp;&nbsp;&nbsp; Improving the steady-state characteristics can be achieved by reducing the steady-state error, which means the system output is much closer to the given input as time goes on. From the steady-state error table, we can say that when the steady-state error is a finite value, as open-loop gain $K_{OL}$ increases, the steady-state error $e_f$ decreases. 

&nbsp;&nbsp;&nbsp; We can also observe that increasing the type number $N$ reduces the steady-state error for each input type. If you increase the type number enough, you can obtain zero steady-state error, $e_f=0$, which means our control system can provide the output equal to the desired input when enough time has passed.

&nbsp;&nbsp;&nbsp; We will continue with a study case, which will show us how we can improve the steady-state characteristics of a feedback control system.

&nbsp;&nbsp;&nbsp; <ins>***Case Study: Improving Steady-State Characteristics of the Feedback DC Servomotor Position Control System***</ins>

&nbsp;&nbsp;&nbsp; In this example, we will observe the steady-state characteristics of the given system by changing its input type and increasing the type number of the open-loop transfer function. To change the type number, we will activate the integrator part of the controller.

<center><img src=https://raw.githubusercontent.com/Dengesizizm/ME462Project/master/Week9/Figures/Example_BD.png height="280"></center>

&nbsp;&nbsp;&nbsp; You can see the block diagram of the servomotor position control system above. The equations of the open-loop transfer function can be written as

$$G_o(s)=G_c(s)G(s)H(s)\quad where \quad G(s) = \frac{100}{s(0.02s+1)}, \qquad H(s)=1 \quad and,$$

$$
  \begin{align*}
 G_c(s)=
  \begin{cases}
   K        & \text{if } \, switch \,\, is \,\, off. \\
  K(1+\frac{1}{Ts})        & \text{if } \, switch \,\, is \,\,on.
  \end{cases}
  \
\end{align*}
$$

&nbsp;&nbsp;&nbsp; Then, the open-loop transfer function when the integrator is not activated can be found as

$$ G_o(s)=\frac{100 K}{s(0.02s+1)} \rightarrow N=1, \quad K_{OL} =100K. $$

&nbsp;&nbsp;&nbsp; When the integrator is activated, open-loop transfer function becomes

$$ G_o(s)=\frac{\frac{100 K}{T}(Ts+1)}{s^2(0.02s+1)} \rightarrow N=2, \quad K_{OL} =\frac{100 K}{T}. $$

&nbsp;&nbsp;&nbsp;  You can check the steady-state error of the given system for different inputs by running the code below. You can also observe the effect of increasing the type number by activating the integrator part of the controller. Since the given system has type number $N=1$ when the integrator is not enabled, the steady-state error can be found as 

$$
  \begin{align*}
 e_f=
  \begin{cases}
   0        & \text{for } step\,\,input. \\
    \frac{1}{100K}        & \text{for } ramp\,\,input. \\
  \infty        & \text{for } acceleration\,\,input.
  \end{cases}
  \
\end{align*}
$$

&nbsp;&nbsp;&nbsp; However, when the integrator part is activated, the type number of the open-loop transfer function increases to $N=2$ and steady-state characteristics of the system improves as

$$
  \begin{align*}
 e_f=
  \begin{cases}
   0        & \text{for } step\,\,input. \\
    0        & \text{for } ramp\,\,input. \\
  \frac{1}{(\frac{100K}{T})}       & \text{for } acceleration\,\,input.
  \end{cases}
  \
\end{align*}
$$

&nbsp;&nbsp;&nbsp; Moreover, you can decrease the steady-state error in the cases when it has a finite value by increasing the open-loop gain $K_{OL}$. You can observe all of these cases by running the code below! 

```
!pip install ipywidgets
!jupyter nbextension enable --py widgetsnbextension
!pip install nodejs
!pip install control
from IPython.display import clear_output
clear_output()

import ipywidgets as widgets
import control
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
from IPython.display import Markdown as md
from control import TransferFunction as tf
from control import minreal
from control import step_response
from IPython.display import clear_output
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, HBox, Label, VBox, interactive_output

def d_e(a):
  a_latex = a._repr_latex_()
  a_latex = a_latex[2:-2]
  return a_latex

def closed_loop(I,S,K,Ti):
  G = tf([100], [0.02,1,0])
  if S == 1:
    Gc = tf([K], [1])
    K_OL = 100*K
    TN=1
  else:
    Gc = tf([K*Ti,K], [Ti,0])
    K_OL = 100*K/Ti
    TN=2
  H = tf([1], [1])
  T = np.linspace(0, 2,1000)
  Gcl = G*Gc/(1+G*Gc*H)
  Go = G*Gc*H
  if I == 1:
    t,y1=step_response(Gcl,T)
    t,y2=step_response(tf([1], [1]),T)
  elif I == 2:
    t,y1=step_response(Gcl*tf([1], [1,0]),T)
    t,y2=step_response(tf([1], [1,0]),T)
  else :
    t,y1=step_response(Gcl*tf([1], [1,0,0]),T)
    t,y2=step_response(tf([1], [1,0,0]),T)
  if S==1 and I==3:
    err = "\infty"
    #err1 = "\infty"
  else :
    err = str(round(abs(y2[999]-y1[999]),4))
    #err1 = str(1/K_OL)
  f = plt.figure(figsize=(17.5,4))
  ax = f.add_subplot(121)
  ax1 = f.add_subplot(122)
  ax.plot(t, y1, "-b", label="C(s)")
  ax.plot(t, y2, "-r", label="R(s)")
  ax.legend(loc="lower right")
  ax1.plot(t, y2-y1, "-g")
  ax.grid()
  ax1.grid()
  ax.set_title('Angular position vs time')
  ax1.set_title('Error vs time')
  ax.set_xlabel('t (s)')
  ax1.set_xlabel('t (s)')
  ax.set_ylabel('θ (rad)')
  ax1.set_ylabel('θ (rad)')
  
  display(md("$$G_o(s) =" + d_e(minreal(Go,verbose=False)) + ",\quad K_{OL} =" + str(K_OL) + ",\quad N =" + str(TN) + ",$$"))
  display(md("$$e_f =" + err + "$$"))
  print("\n")
  
display(md("$ $"))

I = widgets.RadioButtons(options=[('Step', 1), ('Ramp', 2), ('Acceleration', 3)],value=1,description='Input:',disabled=False)
S = widgets.RadioButtons(options=[('Off', 1), ('On', 2)],value=1,description='Integrator:',disabled=False)
K = widgets.FloatSlider(value=5.0, min=0.1,max=10.0, step=0.1,description='K =',disabled=False,continuous_update=False,orientation='horizontal',readout=True,readout_format='.1f',)
Ti = widgets.FloatSlider(value=0.10, min=0.10,max=0.30, step=0.01,description='T =',disabled=False,continuous_update=False,orientation='horizontal',readout=True,readout_format='.2f',)


out = interactive_output(closed_loop,{"I":I,"S":S, "K":K, "Ti":Ti})
hbox1 = HBox((I,S,K,Ti))
display(hbox1,out)
```
