# Transient Response Specifications

&nbsp;&nbsp;&nbsp;  The transient response of the control systems is part of the system's response until it reaches the steady-state. To evaluate the dynamic behavior of a system during its transient phase, *five parameters*, called transient response specifications, are defined based on the *step response* of the system.

&nbsp;&nbsp;&nbsp;  Then, the closed-loop transfer function $G(s)$ and input and output signals $X(s)$ and $Y(s)$ can be seen in the block diagram below

<center><img src="https://raw.githubusercontent.com/Dengesizizm/ME462Project/master/Week11/Part2/Figures/Simple%20Transfer%20Function.png"></img></center>

&nbsp;&nbsp;&nbsp;  where $X(s)=\frac{1}{s}$, $x(t)=1$, is the unit step input and $y_f=\lim_{t\to\infty} y(t) steady-state response of the system.$

<br>

&nbsp;&nbsp;&nbsp; ***1. Delay time, $t_d$:*** It is the minimum time value where the system step response is equal to $0.5y_f$, i.e.,

$$\frac{y(t_d)}{y_f}=0.5.$$
&nbsp;&nbsp;&nbsp; ***2. Rise time, $t_r$:*** It is the minimum time value where the system step response is equal to $y_f$, i.e.,

$$\frac{y(t_d)}{y_f}=1.$$

&nbsp;&nbsp;&nbsp; ***3. Peak time, $t_p$:*** It is the time at system step response reaches to the its peak value, i.e.,

$$y(t_p)=max(y(t)).$$

&nbsp;&nbsp;&nbsp; ***4. Settling time, $t_s$:*** It is the time at when the system unit step response is bounded by a given tolerance band $\varepsilon_s$ ,i.e.,

$$\left|\frac{y(t)-y_f}{y_f}\right|\leq\varepsilon_s\quad \,for\,\,all\quad t\geq t_s.$$

&nbsp;&nbsp;&nbsp; Usually $\varepsilon_s$ given as $0.02$ or $0.05$.

&nbsp;&nbsp;&nbsp; ***5. Maximum overshoot, $\varepsilon_p$:*** It is a measure of how large the maximum value of the unit step response is, i.e.,

$$\varepsilon_p = \frac{max(y(t))-y_f}{y_f}.$$

&nbsp;&nbsp;&nbsp; The maximum overshoot as a percentage $M_p$ can be found as $M_p=100 \varepsilon_p$.

&nbsp;&nbsp;&nbsp; So far, we have learned the definitions of the transient response specifications, and we are ready to work on the treansient response specifications of the second-order systems.

## Transient Response Specifications of the Second Order Systems

&nbsp;&nbsp;&nbsp;  As we know from the previous topics, second-order systems are classified as *underdamped, critically damped and overdamped systems*. We will find the transient response specifications for each case in the following section.



### Overdamped Case ($\xi>1)$

&nbsp;&nbsp;&nbsp;  The transfer function of the overdamped system can be described as

$$G(s)= K\frac{\eta w_n s+ w_n^2}{s^2+2 \xi w_ns+w_n^2} = \frac{K(T_0s+1)}{(T_1s+1)(T_2s+1)}. $$

&nbsp;&nbsp;&nbsp; Then, the ratio of the step response and the steady-state response of the overdamped system can be found as,

$$\frac{y(t)}{y_f}=1-\frac{T_1-T_0}{T_1-T_2}e^{-\frac{t}{T_1}}-\frac{T_0-T_2}{T_1-T_2}e^{-\frac{t}{T_2}}.$$
<br>
<br>

&nbsp;&nbsp;&nbsp; ***i) Peak time, $t_p$***:

&nbsp;&nbsp;&nbsp; Since the slope of the step response is zero at the peak, the peak time can be found as,

$$\frac{\dot{y}(t)}{y_f}=0 \rightarrow \frac{T_1-T_0}{T_1-T_2}\frac{1}{T_1}e^{-\frac{t}{T_1}}+\frac{T_0-T_2}{T_1-T_2}\frac{1}{T_2}e^{-\frac{t}{T_2}} =0. $$

&nbsp;&nbsp;&nbsp; Multiplying by $e^{\frac{t}{T_2}}$ and arranging,

$$ e^{t\big(\frac{T_1-T_2}{T_1T_2}\big)}=\frac{T_1}{T_2}\frac{T_0-T_2}{T_0-T_1}, $$

&nbsp;&nbsp;&nbsp; and taking $ln$ of both sides and solving for $t=t_p$,

$$t_p= \frac{T_1T_2}{T_1-T_2}ln \bigg(\frac{T_1(T_0-T_2)}{T_2(T_0-T_1)} \bigg).$$

&nbsp;&nbsp;&nbsp; Note that $t_p$ is real and positive if $T_1>T_2$ and $\frac{T_1(T_0-T_2)}{T_2(T_0-T_1)}>1 $. 

&nbsp;&nbsp;&nbsp; If $T_0<T_1$ :

$$ T_2(T_0-T_1)\cdot \bigg(\frac{T_1(T_0-T_2)}{T_2(T_0-T_1)}>1\bigg) \rightarrow T_1(T_0-T_2) < T_2(T_0-T_1) \rightarrow (T_1-T_2)T_0<0 \rightarrow T_0<0 \,\,\,where \,\,\, T_1>T_2, $$

&nbsp;&nbsp;&nbsp; shows that $\dot{y}(t_p)<0 $ and $y(t_p)$ is minimum. 

&nbsp;&nbsp;&nbsp; If $T_0>T_1$ :

$$ (T_1-T_2)T_0 >0 \rightarrow T_0 > 0, $$

&nbsp;&nbsp;&nbsp; shows that $\dot{y}(t_p)>0 $ and $y(t_p)$ is maximum if $T_0>T_1>T_2$, there is overshoot. 

<br>
<br>

&nbsp;&nbsp;&nbsp; ***ii) Maximum overshoot, $\varepsilon_p$***:

&nbsp;&nbsp;&nbsp; The maximum overshoot can be found as,

$$\varepsilon_p =\frac{y(t_p)-y_f}{y_f}= \frac{T_0-T_1}{T_1-T_2}\bigg[ \frac{T_2(T_0-T_1)}{T_1(T_0-T_2)} \bigg]^{\frac{T_2}{T_1-T_2}}- \frac{T_0-T_2}{T_1-T_2} \bigg[ \frac{T_2(T_0-T_1)}{T_1(T_0-T_2)} \bigg]^{\frac{T_1}{T_1-T_2}} \,\,\,for\,\,\,T_0>T_1>T_2.$$

<br>

&nbsp;&nbsp;&nbsp; ***iii) Rise time, $t_r$***:

&nbsp;&nbsp;&nbsp; The rise time can be obtained as,

$$\frac{y(t)}{y_f}=1 \rightarrow t_r=\frac{T_1T_2}{T_1-T_2}ln\bigg( \frac{T_0-T_2}{T_0-T_1} \bigg) \,\,\,for\,\,\,T_0>T_1>T_2.$$

<br>

&nbsp;&nbsp;&nbsp; ***iv) Delay time, $t_d$ and settling time, $t_s$***:

&nbsp;&nbsp;&nbsp; These values can be obtained by using their definitions. Their closed-form solutions are not always possible.



### Critically Damped Case ($\xi=1)$

&nbsp;&nbsp;&nbsp;  The transfer function of the critically damped system can be described as

$$G(s)= K\frac{\eta w_n s+ w_n^2}{s^2+2 \xi w_ns+w_n^2} = \frac{K(T_0s+1)}{(Ts+1)^2}. $$

&nbsp;&nbsp;&nbsp; Then, the ratio of the step response and the steady-state response of the overdamped system can be found as,

$$\frac{y(t)}{y_f}=1-\bigg(1+ \frac{T-T_0}{T^2}t\bigg)e^{-\frac{t}{T}} .$$
<br>
<br>

&nbsp;&nbsp;&nbsp; ***i) Peak time, $t_p$***:

&nbsp;&nbsp;&nbsp; Since the slope of the step response is zero at the peak, the peak time can be found as,

$$\frac{\dot{y}(t)}{y_f}=0 \rightarrow -\frac{T-T_0}{T^2}e^{-\frac{t}{T}}+\frac{1}{T} \bigg( 1+ \frac{T-T_0}{T^2}t \bigg) e^{-\frac{t}{T}}=0 \rightarrow e^{-\frac{t}{T}} \bigg(-\frac{T-T_0}{T^2} + \frac{1}{T} + \frac{T-T_0}{T^3}t \bigg) =0 \rightarrow t_p=\frac{T_0T}{T_0-T}, $$
<br>

&nbsp;&nbsp;&nbsp; where,

&nbsp;&nbsp;&nbsp; $t_p>0 \,\,\,if \,\,\, T_0>T \quad \rightarrow \quad y(t_p)=maximum.$

&nbsp;&nbsp;&nbsp; $T_0<0 \qquad \rightarrow \quad y(t_p)=minimum.$

<br>

&nbsp;&nbsp;&nbsp; ***ii) Maximum overshoot, $\varepsilon_p$***:

&nbsp;&nbsp;&nbsp; The maximum overshoot can be found as,

$$\varepsilon_p =\frac{y(t_p)-y_f}{y_f}= \frac{T_0-T}{T}e^{-\frac{T_0}{T_0-T}}  \,\,\,for\,\,\,T_0>T>0.$$


&nbsp;&nbsp;&nbsp; ***iii) Rise time, $t_r$***:

&nbsp;&nbsp;&nbsp; The rise time can be obtained as,

$$\frac{y(t)}{y_f}=1 \rightarrow t_r=\frac{T^2}{T_0-T} \,\,\,for\,\,\,T_0>T>0.$$

### Underdamped Case ($0<\xi<1)$

&nbsp;&nbsp;&nbsp;  The transfer function of the critically damped system can be described as

$$G(s)= K\frac{\eta w_n s+ w_n^2}{s^2+2 \xi w_ns+w_n^2} . $$

&nbsp;&nbsp;&nbsp; Then, the ratio of the step response and the steady-state response of the overdamped system can be found as,

$$\frac{y(t)}{y_f}=1-a_0e^{-\xi w_n t}cos(w_dt-\phi) ,$$

&nbsp;&nbsp;&nbsp; where, 

$$ a_0=\frac{ \sqrt{\eta^2-2\xi\eta+1}}{\sqrt{1-\xi^2}}, \quad \phi=tan^{-1}\bigg( \frac{\xi -\eta}{\sqrt{1- \xi^2}} \bigg)\quad and \quad w_d=w_n \sqrt{1- \xi^2.}$$

<br>
<br>

&nbsp;&nbsp;&nbsp; ***i) Peak time, $t_p$***:

&nbsp;&nbsp;&nbsp; Since the slope of the step response is zero at the peak, the peak time can be found as,

$$\frac{\dot{y}(t)}{y_f}=0 \rightarrow \xi w_na_0e^{-\xi w_nt}cos(w_dt-\phi) +w_da_0e^{-\xi w_n t}sin(w_dt-\phi)=0. $$
<br>

&nbsp;&nbsp;&nbsp; Since $w_na_0e^{-\xi w_n t} \neq 0$, it can be cancelled as

$$\xi cos(w_dt-\phi)+\sqrt{1- \xi^2}sin(w_dt-\phi)=0.$$

&nbsp;&nbsp;&nbsp; From the definition of the damping angle $\beta$, we have

$$\xi = cos\beta \quad and \quad \sqrt{1-\xi^2}=sin\beta. $$

&nbsp;&nbsp;&nbsp; Then, the equation becomes

$$ cos\beta cos(w_dt-\phi)+sin\beta sin(w_dt-\phi))=0 \rightarrow cos(w_dt-\phi - \beta)=0,$$

$$t_p=\frac{\phi + \beta + \frac{\pi}{2}}{w_d} \quad for \quad \eta \neq 0.$$

&nbsp;&nbsp;&nbsp; If $\eta = 0$, then $\phi=tan^{-1} \frac{\xi}{\sqrt{1-\xi^2}}= \frac{\pi}{2}-\beta$, then $t_p$ can be found as

$$t_p=\frac {\pi}{w_d} = \frac{\pi}{w_n \sqrt{1- \xi^2}} \quad for \quad \eta = 0.$$






<br>

&nbsp;&nbsp;&nbsp; ***ii) Maximum overshoot, $\varepsilon_p$***:

&nbsp;&nbsp;&nbsp; The maximum overshoot can be found as,

$$\varepsilon_p =\frac{y(t_p)-y_f}{y_f}= a_0sin\beta\,exp \bigg[ - \frac{\xi}{1-\xi^2} \big( \phi + \beta + \frac{\pi}{2} \big) \bigg] \quad for \quad \eta \neq 0,$$

$$\varepsilon_p = exp \bigg( - \frac{\pi \xi}{\sqrt{1-\xi^2}} \bigg) \quad for \quad \eta = 0.$$

<br>

&nbsp;&nbsp;&nbsp; ***iii) Rise time, $t_r$***:

&nbsp;&nbsp;&nbsp; The rise time can be obtained as,

$$\frac{y(t)}{y_f}=1 \rightarrow t_r=\frac{\phi+\frac{\pi}{2}}{w_d} \quad for \quad \eta \neq 0.$$

&nbsp;&nbsp;&nbsp; If $\eta=0$, then

$$ t_r = \frac{tan^{-1}\bigg( \frac{\xi}{\sqrt{1-\xi^2}} \bigg)+\frac{\pi}{2}}{w_d}.$$


&nbsp;&nbsp;&nbsp; Since $tan^{-1}\bigg( \frac{\xi}{\sqrt{1-\xi^2}} \bigg)=\frac{\pi}{2}-\beta$, it can be obtained that

$$t_r = \frac{\pi-\beta}{w_n\sqrt{1-\xi^2}} \quad for \quad \eta = 0.$$


&nbsp;&nbsp;&nbsp; So far, we obtained the equations of the transient response specifications for second-order systems. We can design the controller with the help of these equations to satisfy desired transient response specifications. Let's look at the following case study to understand how we can design a controller by using the formulas we derived above.

&nbsp;&nbsp;&nbsp; <ins>***Case Study: Proportional Controller Design for the Mass Positioning Servo System***</ins>

 &nbsp;&nbsp;&nbsp; In this example, we will design a proportional controller for the mass positioning servo system to satisfy given transient response requirements. The block diagram of the mass positioning system, including the proportional controller, is given below.

<center><img src=https://raw.githubusercontent.com/Dengesizizm/ME462Project/master/Week11/Part2/Figures/Example.png height="190"></center>

&nbsp;&nbsp;&nbsp; The $F(s)$, $V(s)$ and $X(s)$ signals represents force, velocity and position respectively. The $K_0$ is the controller gain and $T_r$ is the controller time constant. If we take the mass of the object as $m=1kg$, the closed-loop transfer function of the system can be found as,

$$ M(s)=\frac{X(s)}{R(s)}=\frac{K_0}{s^2+K_0T_rs+K_0} .$$

&nbsp;&nbsp;&nbsp; Since we have $T_r \neq 0$, the system is underdamped. Therefore, using the equations we derived for the underdamped system, we can select the proper controller parameters to satisfy transient response requirements. 

 &nbsp;&nbsp;&nbsp; For example, let's assume that you are a control systems engineer, and your customer wants the mass positioning system to have 10 percent overshoot $M_p=10\%$ and 1 second of peak time $t_p=1\,\,s$. To find the required controller parameters we can use the formulas we derived before as

$$M(s)=\frac{K_0}{s^2+K_0T_rs+K_0} = K\frac{\eta w_n s+ w_n^2}{s^2+2 \xi w_ns+w_n^2} = \frac{Kw_n^2}{s^2+2 \xi w_ns+w_n^2} \quad where \quad \eta=0. $$

&nbsp;&nbsp;&nbsp; Then, we can find the damping ratio ($\xi$)  and natural frequency ($w_n$) values from the maximum overshoot ($\varepsilon_p$) and peak time ($t_p$) requirements as

$$\varepsilon_p=exp \bigg( -\frac{\pi \xi}{\sqrt{1-\xi^2}} \bigg)=0.1 \rightarrow \xi=0.591,$$

$$t_p= \frac{\pi}{w_n\sqrt{1-\xi^2}}=1 \rightarrow w_n=3.895\,\,rad/s.$$


&nbsp;&nbsp;&nbsp; By using the damping ratio ($\xi$)  and natural frequency ($w_n$) values, we can find the controller parameters $K_0$ and $T_r$ which satisfies the given transient response conditions as

$$K_0 = w_n^2 \rightarrow K_0=15.172\,\,N/m, $$

$$K_0T_r=2\xi w_n \rightarrow T_r=\frac{2\xi w_n}{K_0}=0.3035\,\,s.$$

&nbsp;&nbsp;&nbsp; We found that if set the controller parameters as $K_0=15.172$ and $T_r=0.3035$, the mass positioning system will have $M_p=100*\varepsilon_p=10\%$ and $t_p=1\,\,s$ which will make your customer happy.

```
!pip install ipywidgets
!jupyter nbextension enable --py widgetsnbextension
!pip install nodejs
!pip install control
from IPython.display import clear_output
clear_output()
import ipywidgets as widgets
import control
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
from IPython.display import Markdown as md
from control import TransferFunction as tf
from control import minreal
from control import impulse_response, step_response
from IPython.display import clear_output
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, HBox, Label, VBox, interactive_output
from control.matlab import *
import warnings

warnings.filterwarnings("ignore")

def rise_time(t,yout):
  i=1
  t_r=0
  while i < 10000:              
    if round(yout[i], 2) == 1:
      t_r = i/1000
      break
    i= i+1
  return yout[i], i-1
   

def d_e(a):
  a_latex = a._repr_latex_()
  a_latex = a_latex[2:-2]
  return a_latex

def closed_loop(K0,Tr):
  G = tf([K0], [1,K0*Tr,K0])
  T1 = np.linspace(0, 10,10000)
  t,y1=step_response(G,T1) 
  e_p = round(100*(max(y1)-1),2)
  t_p = np.argmax(y1)/1000

  t_r,i = rise_time(t,y1)

  i_t = round((i/1000),2)

  plt.plot(t,y1, "-r", label="Step response")
  plt.plot(i/1000,t_r,'*b', label="Rise time", markersize=8.5)
  plt.plot(t_p,max(y1),'*g', label="Peak time", markersize=8.5)
  plt.plot(0, i/1000, linestyle='-')
  plt.plot(0, t_p, linestyle='-')
  plt.grid()
  plt.xlabel('t (s)')
  plt.ylabel('x (m)')
  plt.title("Step response of the Closed-loop system")
  plt.legend(loc="lower right")
  print("\n")

  display(md("$ M(s) =" + d_e(minreal(G,verbose=False)) + ",\quad M_p ="+ str(e_p) +"\, \% ,\quad t_p="+str(t_p)+"s,\quad t_r="+str(i_t)+"s$"))
  print("\n")
display(md("$ $"))
K0 = widgets.BoundedFloatText(value=15.172,min=0.1,max=20.00,step=0.0001,description=u'K0 =',disabled=False,layout=Layout(width='14%', height='80px'),readout=True,readout_format='.4f')
Tr = widgets.BoundedFloatText(value=0.3035,min=0.1,max=0.7500,step=0.0001,description=u'Tr =',disabled=False,layout=Layout(width='14%', height='80px'),readout=True,readout_format='.4f')

hbox1 = HBox((K0,Tr))
out1 = interactive_output(closed_loop,{"K0":K0, "Tr":Tr})

display(hbox1,out1)
```
